```python
import numpy as np 
from PIL import Image, ImageDraw
from matplotlib import pyplot as plt
from matplotlib import gridspec
import load_ship_data
```

```python
# Load the data! 
filePath = './data/shipsnet.json'
trainingSet, testingSet, validationSet = load_ship_data.load_data_train_test_split(filePath)
```

```python
def checkData(dataset): 
    sampleIndex = np.random.randint(len(dataset[0]))
    imageData = dataset[0][sampleIndex]
    hasShip = dataset[1][sampleIndex]
    R = imageData[0]
    G = imageData[1]
    B = imageData[2]
    
    fig = plt.figure(figsize=(8,8), constrained_layout=True)
    spec = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)

    fig_ax1=fig.add_subplot(spec[0,0])
    fig_ax1.set_title('Red')
    fig_ax1.imshow(R, cmap='Reds')
    fig_ax2=fig.add_subplot(spec[0,1])
    fig_ax2.set_title('Green')
    fig_ax2.imshow(G, cmap='Greens')
    fig_ax3=fig.add_subplot(spec[1,0])
    fig_ax3.set_title('Blue')
    fig_ax3.imshow(B, cmap='Blues')
    fig_ax4=fig.add_subplot(spec[1,1])
    fig_ax4.set_title('Red')
    fig_ax4.imshow(imageData.transpose())
    
    fig.suptitle(f"Sampled RGB And Image. Ship Exists:{hasShip}")

checkData(trainingSet)
```
